### Объектно-ориентированное программирование (ООП)

ООП - парадигма программирования, основанная на концепциях объектов и классов.

*   Класс — тип, описывающий устройство объектов. 
*   Объект — экземпляр класса. Класс - основа, на которой определяются объекты.

В python всё является объектами (строки, списки, ...). Например, строка str() - это экземпляр класса строк или, иначе говоря, объект класса строк.

Но возможности ООП в python этим не ограничены. Мы можем написать свой тип данных (класс), определить в нём свои методы.

Это не является обязательным - мы можем пользоваться только встроенными объектами. Однако ООП полезно при долгосрочной разработке программы несколькими людьми, оно сильно упрощает понимание кода.


Стандартная конструкция 


```
class [название класса]() 
```

Для демонстрации ООП будем решать простую задачу:

Создадим 2 функции, которые назовем соответственно earn и spend.

* Функция earn берет начальное богатство потребителя w и добавляет к нему свой текущий заработок y.

* Функция spend берет начальное богатство потребителя w и вычитает из нее свои текущие расходы x.

````python
def earn(w, y):
    "Человек с начальным капиталом w зарабатывает y"
    return w + y

def spend(w, x):
    "Человек с начальным капиталом w тратит x"
    new_wealth = w - x
    if new_wealth < 0:
        print("Недостаточно средств")
    else:
        return new_wealth
````

````python
# Мы можем использовать эти две функции для отслеживания состояния потребителя по мере того, как он зарабатывает и тратит.
# Например:

w0 = 100
w1 = earn(w0,  10)
w2 = spend(w1, 20)
w3 = earn(w2,  10)
w4 = spend(w3, 20)
print(f"w0 = {w0}, w1 = {w1}, w2 = {w2}, w3 = {w3}, w4 = {w4}")
````

Класс связывает набор данных, связанных с конкретным экземпляром, вместе с набором функций, которые работают с данными.

В нашем примере экземпляром будет имя конкретного человека, данные экземпляра которого состоят исключительно из его состояния/начального капитала.

В нашем примере к данным текущего экземпляра можно применить две функции: **earn** и **spend**. Данные экземпляра и функции называются методами.
К ним можно легко получить доступ способами, которые мы сейчас опишем.

Мы построим Consumer класс, включив в него :

- атрибут капитала, в котором хранятся данные о капитале потребителя

- метод **earn**, где заработок (у) увеличивает благосостояние потребителя на у

- метод **spend**, где трата (х) либо уменьшает богатство на х, либо возвращает ошибку, если средств недостаточно

````python
class Consumer:
    def __init__(self, w):
        # __init__ - конструктор класса
        # после того, как класс инициализирован - в него можно принимать агрументы
        # конструкция __init__ запускается автоматически
        # Вызов __init__ устанавливает «пространство имен» для хранения данных экземпляра.
        
        "Инициализируем потребителя с начальным капиталом w долларов"
        self.wealth = w
        self.age = 30

    def earn(self, y):
        "Потребитель зарабатывает y долларов"
        self.wealth = self.wealth + y

    def spend(self, x):
        "Потребитель тратит х долларов, если это возможно"
        new_wealth = self.wealth - x
        if new_wealth < 0:
            print("Недостаточно средств")
        else:
            self.wealth = new_wealth
            
````

**Зачем нам self?**  
Классам нужен способ, что ссылаться на самих себя т.е. способ сообщения между экземплярами. Слово **self** это способ описания любого объектов.

Здесь есть особый синтаксис, поэтому давайте внимательно рассмотрим его:

- Ключевое слово class указывает, что мы создаем класс.

- Класс Consumer определяет начальные данные о капитале (`w`) и три метода: `__init__`, `earn` и `spend`.

- Капитал `w` - это данные экземпляра, потому что каждый созданный нами потребитель (каждый экземпляр класса Consumer) будет иметь свои собственные данные о капитале.

- Метод `__init__` - это конструктор класса. Всякий раз, когда мы создаем экземпляр класса, метод `__init__` будет вызываться автоматически. Вызов `__init__` устанавливает «пространство имен» для хранения данных экземпляра.

````python
# Пример, в котором мы используем класс Consumer для создания экземпляра потребителя, которого мы назовем c1.
# Создав потребителя c1 и наделив его начальным капиталом 10, мы применим метод затрат.

c1 = Consumer(10)

#  Здесь обанкротим его
c1.spend(10)
print(c1.wealth)
c1.spend(1)
c1.earn(10)
print(c1.wealth)
````

````python
# Место для вопросов и написания класса банокомата ATM

class ATM():
    def __init__(self, d):
        """
        d = {Номинал купюры: число купюр} 10, 50, 100
        """
        self.d = d

    def get_money(self, user_dict): # {10: 2, 50:1, 100:1}
        if self.control(user_dict) == False:
            print('Денег нет')
            return
        
        for money in user_dict:
            self.d[money] -= user_dict[money]

        return user_dict

    def take_money(self, user_dict): # {10: 2, 50:1, 100:1} 
        for key in user_dict.keys():   # key = 10, key = 50, key = 100
            self.d[key] += user_dict[key]

    def control(self, user_dict):
        """
        return: True -- если купюры есть, False иначе
        """
        for key in user_dict.keys():   # key = 10, key = 50, key = 100
            if self.d[key] < user_dict[key]:
                return False
        return True
````

````python
atm = ATM({10: 20, 50: 10, 100: 2})
atm.get_money({10:1})
print(atm.d)
atm.get_money({100:100})
atm.take_money({100:100})
print(atm.d)
````

### Magic methods и перегрузка операторов

Вы, наверное, обратили внимание на то, как выглядит метод `__init__`, вызывающий конструктор класса. Такие методы, название которых начинается и заканчивается двумя нижними подчеркиваниями, носят название **магических методов** (magic methods). И с помощью них действительно можно творить всякую магию:)

Например, определяя такие методы для наших классов, можно выполнить **перегрузку операторов**, т. е. переопределять привычные операторы (+, -, * и другие) так, чтобы они работали для наших классов именно таким образом, как мы этого хотим. Пример: для класса `list` оператор сложения объединяет два списка в один, а для `numpy.ndarray` сложение --- это именно поэлементное сложение массивов.

Посмотреть список магических методов можно [здесь](https://rszalski.github.io/magicmethods/)

Давайте сделаем свой класс строк, в котором определим операцию вычитания одной строки из другой следующим образом: строка a минус строка b -- это строка a, из которой мы убрали первое вхождение b, если оно есть. Если его нет, то строка a останется без изменения

````python
class StrWithSubtraction:
  def __init__(self, s):
    self.string = s # содержимое строки

  def __add__(self, other_s): # сложение аналогично сложению в str
    self.string = self.string + other_s 
    return self.string

  def __sub__(self, other_s): # определение оператора вычитания
    first_occurence = self.string.find(other_s)
    if first_occurence == -1: # если не нашли, то возвращаем исходную строку
      return self.string
    else:
      self.string = self.string[:first_occurence] + self.string[first_occurence+len(other_s):] # берем всю строку, кроме фрагмета, содержащего other_s
      return self.string
````
